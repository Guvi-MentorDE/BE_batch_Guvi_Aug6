```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
data = pd.read_csv('D:\Guvi\dashboard\case_time_series.csv')
```

```python
Y = data.iloc[61:,1].values 
R = data.iloc[61:,3].values 
D = data.iloc[61:,5].values 
X = data.iloc[61:,0] 
```

```python
plt.figure(figsize=(25,8))
 
ax = plt.axes()
ax.grid(linewidth=0.4, color='#8f8f8f') 
 
ax.set_facecolor("black") 
ax.set_xlabel('\nDate',size=25,color='#4bb4f2')
ax.set_ylabel('Number of Confirmed Cases\n',
              size=25,color='#4bb4f2')
#‘.xticks’ and ‘.yticks’ lets you alter the Dates and Daily Confirmed font. 
# For the Dates to not overlap with each other we are going to represent them vertically by passing ‘rotation = vertical’. 

#‘.tick_params’ lets you alter the size and color of the dashes which look act like a bridge between the ticks and graph.
plt.xticks(rotation='vertical',size='20',color='white')
plt.yticks(size=20,color='white')
plt.tick_params(size=20,color='white')

#The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together etc.

for i,j in zip(X,Y):
    ax.annotate(str(j),xy=(i,j+100),color='white',size='13')
     
ax.annotate('Second Lockdown 15th April',
            xy=(15.2, 860),
            xytext=(19.9,500),
            color='white',
            size='25',
            arrowprops=dict(color='red',
                            linewidth=0.025))
 
plt.title("COVID-19 INDIA : Daily Confirmed\n",
          size=50,color='blue')
 
ax.plot(X,Y,
        color='#1F77B4',
        marker='o',
        linewidth=4,
        markersize=15,
        markeredgecolor='#035E9B')
```
